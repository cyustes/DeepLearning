# DATA ANALYSIS

```python
import pandas as pd

#Dataset
file_path = 'raw_sales.csv'
data = pd.read_csv(file_path)

# Display of the first few rows of the dataset
data.head()
```

```python
data
```

```python
# Preprocess
data = pd.read_csv(file_path, sep=',')

#'datesold' column to datetime format
data['datesold'] = pd.to_datetime(data['datesold'])

data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#plots
sns.set(style="whitegrid")

# Exploratory Data Analysis

# Plotting price over time
plt.figure(figsize=(12, 6))
plt.plot(data['datesold'], data['price'], marker='o', linestyle='', alpha=0.5)
plt.title('House Prices Over Time')
plt.xlabel('Date Sold')
plt.ylabel('Price')
plt.show()

# Analyzing the distribution of prices
plt.figure(figsize=(12, 6))
sns.histplot(data['price'], kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Analyzing the relationship between price and number of bedrooms
plt.figure(figsize=(12, 6))
sns.boxplot(x='bedrooms', y='price', data=data)
plt.title('House Prices by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()

# Analyzing the relationship between price and property type
plt.figure(figsize=(12, 6))
sns.boxplot(x='propertyType', y='price', data=data)
plt.title('House Prices by Property Type')
plt.xlabel('Property Type')
plt.ylabel('Price')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filtering out the entries where the number of bedrooms is 0
filtered_data = data[data['bedrooms'] != 0]

# Boxplot with the filtered data
plt.figure(figsize=(12, 6))
sns.boxplot(x='bedrooms', y='price', data=filtered_data)
plt.title('House Prices by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

```python
# Correlation analysis between house prices and postcodes
# Median house prices per postcode
median_prices_by_postcode = data.groupby('postcode')['price'].median().sort_values()

# Median house prices per postcode
plt.figure(figsize=(12, 6))
median_prices_by_postcode.plot(kind='bar')
plt.title('Median House Prices by Postcode')
plt.xlabel('Postcode')
plt.ylabel('Median Price')
plt.show()


# Box plot price and bedrooms
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[['price', 'bedrooms']])
plt.title('Boxplot for Price and Bedrooms')
plt.show()
```
